```python
#Llegim el fitxer amb les dades preprocessades
import numpy as np
import pandas as pd

fitxer="D:/Users/X/Downloads/smoking_drinking_dataset_pre.csv"
df=pd.read_csv(fitxer,encoding='utf-8',sep=',')
df.head()
```

```python
#Veiem la distribució d'observacions per edat i per genere
#podem veure que la majoria d'observacions están entre individus de 30 i 60 anys i
#que la diferència per géneres no és molt accentuada

#Un efecte interessant que s'observa és que en edats menors a 50 anys hi ha mes homes que dones
#mentre que en edats mes avançades s'equilibren els valors inclós tendeixen a superar les dones als homes

import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme()
plt.figure(figsize=(12,6))
sns.histplot(data=df,x='edat',hue='genere',multiple='dodge',shrink=5) 
#plt.legend(labels=['Home','Dona'])
```

```python
#Veiem la distribució entre sexes del nivell de fumador

#Podem veure que les mostres de les dones son majoritariament de no fumadors 
#mentre que els homes es distribueixen entre les categories de no fumadors, 
#exfumadors i fumadors
#El volum de dones fumadores és sensiblement més baix que el de homes fumadors

print(df[['genere','nivell_fumador']].value_counts())
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
sns.violinplot(data=df, x='genere', y='nivell_fumador_num',ax=axes[0])
sns.countplot(x=df['nivell_fumador'],hue=df['genere'],ax=axes[1])
#labels=['No fumador','ExFumador','Fumador']
#axes[1].set_xticklabels(labels, rotation=0)
#axes[1].legend(labels=['Home','Dona'])
```

```python
#Veiem la distribucio entre sexes de la categoria bevedor
#Podem veure que hi ha mes volum d'homes bevedors que de dones

print(df[['genere','bevedor']].value_counts())
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
sns.violinplot(data=df, x='genere', y='bevedor_num',ax=axes[0])
xlabels=['Home','Dona']
axes[0].set_xticklabels(xlabels, rotation=0)
#ylabels=['No  bevedor','Bevedor']
#axes[0].set_yticklabels(ylabels, rotation=0)
sns.countplot(x=df['bevedor'],hue=df['genere'],ax=axes[1])
#labels=['No bevedor','Bevedor']
#axes[1].set_xticklabels(labels, rotation=0)
#axes[1].legend(labels=['Home','Dona'])
```

```python
# Veiem la relación entre els habits de fumar i de beure, en relació a la població, amb un mapa de calor.
# Això ens permetra veure, de forma diferenciada per homes i dones, les diferents concentracions de
# poblacio que tenen aquests habits

# Diferenciem homes i dones ja que hem vist anteriormentt que aquest habits es comporten de forma diferent
# en uns i altres i entenem que separant-los coneixerem millor el nostre dataset

#  La major concentracio de homes es troba en el grup de fumadors i no bevedors
# mentre que la major concentració de dones es troba en el grup no fumadores i no bevedores

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

df_h=df[df['genere']=='Home'][['nivell_fumador','bevedor']] #home
df_d=df[df['genere']=='Dona'][['nivell_fumador','bevedor']] #dona
#print(df_h)
df_h_group=df_h.groupby(['nivell_fumador','bevedor']).size().reset_index(name='counts')
#print('Taula agrupada\n',df_h_group)

#create pivot table per genere Homes
df_pivot_h = pd.pivot_table(df_h_group, values='counts', index='nivell_fumador', columns='bevedor',
                          aggfunc='sum')
print('Taula Homes\n',df_pivot_h)
sns.heatmap(data=df_pivot_h,annot=True,fmt=".0f",cmap='coolwarm',ax=axes[0]).set_title('Homes')

#print(df_h)
df_d_group=df_d.groupby(['nivell_fumador','bevedor']).size().reset_index(name='counts')
#print(df_d_group)

#create pivot table per genere Dones
df_pivot_d = pd.pivot_table(df_d_group, values='counts', index='nivell_fumador', columns='bevedor',
                          aggfunc='sum')
print('Taula Dones\n',df_pivot_d)
sns.heatmap(data=df_pivot_d,annot=True,fmt=".0f",cmap='coolwarm',ax=axes[1]).set_title('Dones')


#sns.scatterplot(data=df_h,x="nivell_fumador",y="PAS",ax=axes[0])
#sns.scatterplot(data=df_d,x="nivell_fumador",y="PAS",ax=axes[1])


#sns.swarmplot(data= df, x= 'nivell_fumador', y = 'bevedor', hue= 'genere', dodge=True);
```

```python
#Tensió arterial en relacio a l'habit de beure

#Incrementa amb l'edat, independentment de l'habit de beure, mantenint-se lleugerament més alta en el 
#cas de bevedors

#A considerar el canvi de comportament de la PAD en el cas dels homes a partir dels 50 anys

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
dataset=pd.DataFrame(df[['genere','edat','PAS','PAD','bevedor']])
datasetMean=pd.DataFrame(dataset.groupby(['genere','edat','bevedor']).mean(''))
datasetMean=datasetMean.reset_index()
#print(datasetMean)
datasetMeanHome=datasetMean[datasetMean['genere']=='Home']
datasetMeanDona=datasetMean[datasetMean['genere']=='Dona']
sns.lineplot(data=datasetMeanHome,x='edat',y='PAS',hue='bevedor',ax=axes[0,0])
axes[0,0].set_title('Homes')
axes[0,1].set_title('Homes')
sns.lineplot(data=datasetMeanHome,x='edat',y='PAD',hue='bevedor',linestyle = "dashed",ax=axes[0,1])
axes[1,0].set_title('Dones')
axes[1,1].set_title('Dones')
#axes[0,0].legend(labels=['No bevedor','Bevedor'])
#axes[0,1].legend(labels=['No bevedor','Bevedor'])
sns.lineplot(data=datasetMeanDona,x='edat',y='PAS',hue='bevedor',ax=axes[1,0])
sns.lineplot(data=datasetMeanDona,x='edat',y='PAD',hue='bevedor',linestyle = "dashed",ax=axes[1,1])
#axes[1,0].legend(labels=['No bevedor','Bevedor'])
#axes[1,1].legend(labels=['No bevedor','Bevedor'])
```

```python
#Tensió arterial en relació a l'habit de fumar per edat

# Com en el cas del beure veiem que la pressió arterial incrementa amb l'edat amb la peculiaritat
# de la PAD en el cas dels homes qu decementa a partir del 50

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
dataset=pd.DataFrame(df[['genere','edat','PAS','PAD','nivell_fumador']])
datasetMean=pd.DataFrame(dataset.groupby(['genere','edat','nivell_fumador']).mean(''))
datasetMean=datasetMean.reset_index()
#print(datasetMean)
datasetMeanHome=datasetMean[datasetMean['genere']=='Home']
datasetMeanDona=datasetMean[datasetMean['genere']=='Dona']
axes[0,0].set_title('Homes')
axes[0,1].set_title('Homes')
sns.lineplot(data=datasetMeanHome,x='edat',y='PAS',hue='nivell_fumador',ax=axes[0,0])
sns.lineplot(data=datasetMeanHome,x='edat',y='PAD',hue='nivell_fumador',linestyle = "dashed",ax=axes[0,1])
axes[1,0].set_title('Dones')
axes[1,1].set_title('Dones')
#axes[0,0].legend(labels=['No fumador','ExFumador','Fumador'])
#axes[0,1].legend(labels=['No fumador','ExFumador','Fumador'])
sns.lineplot(data=datasetMeanDona,x='edat',y='PAS',hue='nivell_fumador',ax=axes[1,0])
sns.lineplot(data=datasetMeanDona,x='edat',y='PAD',hue='nivell_fumador',linestyle = "dashed",ax=axes[1,1])
#axes[1,0].legend(labels=['No fumador','ExFumador','Fumador'])
#axes[1,1].legend(labels=['No fumador','ExFumador','Fumador'])
```

```python
# Matriu correlació
# Volem veure de una manera ràpida l posible interrelació entre les variables numèriques del nostre dataset.
# Per fer-ho calculaerem la matriu de correlació i la visualitzarem amb un mapa de calor, remarcant aquells valors
# que indiquin, de forma directa o inversa, una forta relació entre els atributs.


import matplotlib.pyplot as plt
import seaborn as sns


#eliminem les columnes no numeriques
dataset=df.drop(['genere','nivell_fumador','bevedor',
                 'oide_esquerra','oide_dret','genere_num',
                 'bevedor_num','nivell_fumador_num','proteina_orina'],axis=1)

mc=dataset.corr()
#fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
#plt.figure(figsize=(10, 10))
mc
```

```python
#mapa de calor
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
sns.heatmap(mc,cmap='Blues',vmin=-1,vmax=1)
```

```python
# resaltem els valors mes significatius per facilitar la identificació
# de variable correlacionades

#A primera ista sembla que les variables relacionades són força evidents:
#- pes i alçada
#- mida cintura i pes
#- visió dreta i visió esquerra
#- tensió arterial alta i baixa
#- oide dret i oide esquerra
#- Nivells de transaminases
#- hemoglobina i alçada (correlació lleu però curiosa)
# Al tractar-se de valors positius, són relacions directes

mcF=mc[((mc >= .5) | (mc <= -.5)) & (mc !=1.000)]
plt.figure(figsize=(20,10))
sns.heatmap(mcF, annot=True, cmap="Reds",annot_kws={'size': '20'})
#sns.set(font_scale=2)
#plt.show()
```
