```python
# Carreguem les dades en un dataframe a partir d'un fitxer csv
import pandas as pd

fitxer="D:/Users/X/Downloads/smoking_driking_dataset_Ver01.csv"
df=pd.read_csv(fitxer,encoding='utf-8',sep=',')
df.head()
```

```python
# Registres duplicats
print('registres duplicats:',df.duplicated().sum())

df[df.duplicated()] 

#Hi ha registres registres duplicats pero no considerem necessari eliminar-los per les següents raons:
#- No representen un volum significatiu
#- Són dades comuns sobre indicadors de salut
#- No hi han dades que permetin identificar únivocament les registres per ser considerades sensibles o de caràcter privat,
# per tant no podem deduir si els duplicats pertanyen al mateix individu
```

```python
# Adequació de columnes
# Canviem el nom de les columnes per disposar de noms més familiars
# Afegim columnes numèriques per representar els atributs categòrics que necessitarem en alguns models 
# que emprarem. Mantenim la columna categòrica descritiva.

print(df.columns)
df=df.rename(columns={'sex':'genere', 'age':'edat', 'height':'alcada', 'weight':'pes', 'waistline':'cintura', 'sight_left':'visio_esquerra',
       'sight_right':'visio_dreta', 'hear_left':'oide_esquerra_num', 'hear_right':'oide_dret_num', 'SBP':'PAS', 'DBP':'PAD', 'BLDS':'NGS',
       'tot_chole':'colesterol_total', 'HDL_chole':'colesterol_HDL', 'LDL_chole':'colesterol_LDL', 'triglyceride':'triglicerid', 'hemoglobin':'hemoglobina','urine_protein':'proteina_orina', 'serum_creatinine':'creatinina', 'SGOT_AST':'transaminasa_AST', 'SGOT_ALT':'transaminasa_ALT',
       'gamma_GTP':'transferasa_GGT', 'SMK_stat_type_cd':'nivell_fumador_num', 'DRK_YN':'bevedor'})
#df.rename(...,inplace=True)
print(df.columns)
#canviem valors categorics per numèrics
df['genere'] = df['genere'].str.replace('Male','Home')
df['genere'] = df['genere'].str.replace('Female','Dona')
df['bevedor'] = df['bevedor'].str.replace('Y','Si')
df['bevedor'] = df['bevedor'].str.replace('N','No')
df['nivell_fumador']=df['nivell_fumador_num']
df['nivell_fumador']=df['nivell_fumador'].astype(str)
df['nivell_fumador'] = df['nivell_fumador'].str.replace("1.0",'No fumador',regex=False)
df['nivell_fumador'] = df['nivell_fumador'].str.replace("2.0",'ExFumador',regex=False)
df['nivell_fumador'] = df['nivell_fumador'].str.replace("3.0",'Fumador',regex=False)
df['oide_esquerra']=df['oide_esquerra_num']
df['oide_esquerra']=df['oide_esquerra'].astype(str)
df['oide_esquerra'] = df['oide_esquerra'].str.replace("1.0",'Normal',regex=False)
df['oide_esquerra'] = df['oide_esquerra'].str.replace("2.0",'No normal',regex=False)
df['oide_dret']=df['oide_dret_num']
df['oide_dret']=df['oide_dret'].astype(str)
df['oide_dret'] = df['oide_dret'].str.replace("1.0",'Normal',regex=False)
df['oide_dret'] = df['oide_dret'].str.replace("2.0",'No normal',regex=False)
df
```

```python
#Transformen els atributs categòrics en numèrics

from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
enc.fit(df[["genere","bevedor"]])
df[["genere_num","bevedor_num"]] = enc.transform(df[["genere","bevedor"]])
print(enc.categories_)
df
```

```python
#convertim atributs float en int
df['genere_num']=df['genere_num'].astype('int64')
df['nivell_fumador_num']=df['nivell_fumador_num'].astype('int64')
df['bevedor_num']=df['bevedor_num'].astype('int64')
df['oide_esquerra_num']=df['oide_esquerra_num'].astype('int64')
df['oide_dret_num']=df['oide_dret_num'].astype('int64')
df['proteina_orina']=df['proteina_orina'].astype('int64')
df.dtypes
```

```python
#guardem el fitxer amb les dades preparades
fitxer="D:/Users/X/Downloads/smoking_drinking_dataset.csv"
df.to_csv(fitxer,encoding='utf-8',sep=',',index=False)
```
